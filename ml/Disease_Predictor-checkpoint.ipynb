```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("improved_disease_dataset.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df['disease'].unique()
```

```python
df.isnull().sum()
```

```python
encoder=LabelEncoder()
```

```python
df['disease']=encoder.fit_transform(df['disease'])#the diseases are now converted to numbers 
```

```python
df['disease'].unique()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(18,8))
sns.countplot(x=y)
plt.title("Disease Class Distribution before Sampling")
plt.xticks(rotation=90)
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler #The dataset which holds less value is duplicated ensuring all have equal samples for fair and effective model training.
ros=RandomOverSampler(random_state=42)
x_sampled,y_sampled=ros.fit_resample(x,y)
```

```python
x_sampled
```

```python
y_sampled
```

```python
plt.figure(figsize=(18,8))
sns.countplot(x=y_sampled)
plt.title("Disease Class Distribution After Sampling")
plt.xticks(rotation=90)
plt.show()
```

```python
if 'gender' in x_sampled.columns:
    le=LabelEncoder()
    x_sampled['gender']=le.fit_transform(x_sampled['gender'])
```

```python
x_sampled=x_sampled.fillna(0)
```

```python
x_sampled
```

```python
if len(y_sampled) > 1:
    y_sampled=y_sampled.values.ravel()
```

```python
from sklearn.tree import DecisionTreeClassifier
models={
    "DT":DecisionTreeClassifier(),
    "RF":RandomForestClassifier(),
    "svm":SVC()
}
```

```python
from sklearn.model_selection import StratifiedKFold


```

```python
cv_scoring='accuracy'
skf=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for model_name ,model in models.items():
    try:
        scores=cross_val_score(model,x_sampled,y_sampled,cv=skf,scoring=cv_scoring,n_jobs=-1,error_score='raise')

        print("=" *50)
        print(f"Model:{model_name}")
        print(f"Scores{scores}")
        print(f"Mean Accuracy:{scores.mean():.4f}")
    except Exception as e:
        print("=" * 50)
        print(f"Model:{model_name} failed with error")
        print(e)
        
```

```python
svm=SVC()
svm.fit(x_sampled,y_sampled)
svm_preds=svm.predict(x_sampled)
cf_matrix=confusion_matrix(y_sampled,svm_preds)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(cf_matrix,annot=True,fmt="d")
plt.title("Confusion Matrix for SVM Classifier")
plt.show()
print(f"SVM Accuracy :{accuracy_score(y_sampled,svm_preds)*100:.2f}%")
```

```python
nb=GaussianNB()
nb.fit(x_sampled,y_sampled)
nb_preds=nb.predict(x_sampled)
cf_matrix=confusion_matrix(y_sampled,nb_preds)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(cf_matrix,annot=True,fmt="d")
plt.title("Confusion Matrix for Naive Bayes  Classifier")
plt.show()
print(f"Gaussian  :{accuracy_score(y_sampled,svm_preds)*100:.2f}%")
```

```python
rfc=RandomForestClassifier(random_state=42)
rfc.fit(x_sampled,y_sampled)
rfc_preds=rfc.predict(x_sampled)
cf_matrix=confusion_matrix(y_sampled,rfc_preds)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(cf_matrix,annot=True,fmt="d")
plt.title("Confusion Matrix for Naive Bayes  Classifier")
plt.show()
print(f"Random Forest Classifier  :{accuracy_score(y_sampled,rfc_preds)*100:.2f}%")
```

```python
from statistics import mode
```

```python
final_preds=[mode([i,j,k]) for i,j,k in zip(svm_preds,nb_preds,rfc_preds)]
cf_matrix_combined=confusion_matrix(y_sampled,final_preds)
plt.figure(figsize=(12,8))
sns.heatmap(cf_matrix_combined,annot=True,fmt="d")
plt.title("Confusion Matrix for Combined Model")
plt.show()
print(f"Combined Accuracy :{accuracy_score(y_sampled,rfc_preds)*100:.2f}%")
```

```python
symptoms = x.columns.values
symptom_index = {symptom: idx for idx, symptom in enumerate(symptoms)}

def predict_disease(input_symptoms):
    input_symptoms = input_symptoms.split(",")
    input_data = [0] * len(symptom_index)
    
    for symptom in input_symptoms:
        if symptom in symptom_index:
            input_data[symptom_index[symptom]] = 1
    
    input_data = np.array(input_data).reshape(1, -1)

    rf_pred = encoder.classes_[rfc.predict(input_data)[0]]
    nb_pred = encoder.classes_[nb.predict(input_data)[0]]
    svm_pred = encoder.classes_[svm.predict(input_data)[0]]

    final_pred = mode([rf_pred, nb_pred, svm_pred])
    
    return {
        "Random Forest Prediction": rf_pred,
        "Naive Bayes Prediction": nb_pred,
        "SVM Prediction": svm_pred,
        "Final Prediction": final_pred
    }

print(predict_disease("Itching,Skin Rash,Nodal Skin Eruptions"))
```
